```python
import numpy as np
import pandas as pd 
```

```python
train_data = pd.read_csv('./train.csv')
test_data = pd.read_csv('./test.csv')
```

```python
train_data
```

```python
#根据数据表述去除无关特征
train_data = train_data.drop(['Name','Ticket','Cabin'],axis=1)
test_data = test_data.drop(['Name','Ticket','Cabin'],axis=1)
```

```python
test_data.info()
```

```python
#缺失数据填充
train_data["Age"] = train_data["Age"].fillna(train_data["Age"].mean())
train_data["Embarked"] = train_data["Embarked"].fillna(method='bfill')
test_data["Age"] = test_data["Age"].fillna(test_data["Age"].mean())
test_data["Fare"] = test_data["Fare"].fillna(test_data["Fare"].mean())
```

```python
from sklearn.preprocessing import LabelEncoder
#将类别变量转为数值
enc = LabelEncoder()
train_data['Sex'] = enc.fit_transform(train_data['Sex'])
test_data['Sex'] = enc.transform(test_data['Sex'])
enc_embarked = LabelEncoder()
train_data['Embarked'] = enc_embarked.fit_transform(train_data['Embarked'])
test_data['Embarked'] = enc_embarked.transform(test_data['Embarked'])
```

```python
test_data
```

```python
from sklearn.preprocessing import StandardScaler
scaler_age = StandardScaler()
train_data['Age'] = scaler_age.fit_transform(np.array(train_data['Age']).reshape(-1,1))
test_data['Age'] = scaler_age.transform(np.array(test_data['Age']).reshape(-1,1))

scaler_fare = StandardScaler()
train_data['Fare'] = scaler_fare.fit_transform(np.array(train_data['Fare']).reshape(-1,1))
test_data['Fare'] = scaler_fare.transform(np.array(test_data['Fare']).reshape(-1,1))
```

```python
train_data
```

```python
x_train = train_data.drop(['PassengerId','Survived'],axis=1)
y_train = train_data['Survived']
x_test = test_data.drop(['PassengerId'],axis=1)
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
import xgboost
import catboost
```

```python
y_train
```

```python
kf = KFold(n_splits=5, random_state=2,shuffle=True)
i = 1 
x_train = np.array(x_train)
y_train = np.array(y_train)
for train_index,val_index in kf.split(x_train,y_train):
    print('\n{} of kfold {}'.format(i,kf.n_splits))
    X_train,X_val = x_train[train_index],x_train[val_index]
    Y_train,Y_val = y_train[train_index],y_train[val_index]
#     model = xgboost.XGBClassifier(max_depth=4,
#                     learning_rate=0.01,
#                     n_estimators=250, # 使用多少个弱分类器
#                     objective='binary:logistic',
#                     booster='gbtree',
#                     gamma=0,
#                     min_child_weight=1,
#                     max_delta_step=0,
#                     subsample=1,
#                     colsample_bytree=1,
#                     reg_alpha=0,
#                     reg_lambda=1,
#                     seed=4 # 随机数种子
#                                  )
    model = catboost.CatBoostClassifier(iterations=250,
                                        learning_rate=0.01,
                                        depth=4,
                                        loss_function='CrossEntropy',
                                        custom_metric='Accuracy',
                                        eval_metric='Accuracy',
                                        leaf_estimation_method='Gradient',
                                        random_seed=8,
                                        verbose=0)
    model.fit(X_train,Y_train)
    pred_result = model.predict(X_val)
    score = accuracy_score(Y_val,pred_result)
    print("准确率：",score)
    i +=1
```

```python
y_test = model.predict(x_test)
```

```python
y_test
```

```python
result = pd.DataFrame(columns=['PassengerId','Survived'])
result['PassengerId'] = test_data['PassengerId']
result['Survived'] = pd.DataFrame(y_test)
```

```python
result.to_csv("./result.csv",index=False)
```
